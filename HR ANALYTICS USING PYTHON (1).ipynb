ATTRITION BY DEPARTMENT

comparing departments
Human resources department has the lower number of people leaving with research department amounting to about 150 employees turnover rate 
however this can not be used to measure the replacement level because if we have 800 employees in research dept and 150 leaves we still have about 650 employees in the department while if sales has 350 employees and 100 leaves we are left with 250 meaning the dept to fill in the most should be sales. 

```python
plt.figure(figsize=(12,5))
sns.countplot(x='Department',hue='Attrition', data=dataset, palette='hot')
plt.title("Attrition w.r.t Department")
plt.show()
```

```python
# impact of travel on Attrition
sns.countplot(hue=dataset.Attrition,x=dataset.BusinessTravel)
plt.show()
```

```python
.from the above Graph it show us that the company has more number of people who travel rarely
.The numbers of employees that travelrarely have more attrition yes meaning they leave the company more and arent satisfied with their jobs
.Those who dont travel at all have less attrition.
```

EDUCATION LEVEL BY ATTRITION


```python
edu_map = {1 :'Below College', 2: 'College', 3 :'Bachelor', 4 :'Master', 5: 'Doctor'}
plt.figure(figsize=(12,5))
sns.countplot(x=dataset['Education'].map(edu_map), hue='Attrition', data=dataset, palette='hot')
plt.title("Education w.r.t Attrition")
plt.show()
```

```python
OBSERVATION
Employees from bachelors  leave  more 
```

```python
General observation
From the dataset it was observed that the most factor influencing Attrition is overtime , when people work overtime there are higher chances of them leaving 
singles leave more than married and divorced. 
male gender also leave more than female
```
